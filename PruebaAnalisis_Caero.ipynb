```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
df
```

```python
df.isnull().sum
```

```python
df.info() #No tiene ob jetos vacio
```

```python
counts = df['Air_Quality_Index'].value_counts()
counts.plot(kind='bar')

plt.title('Calidad del aire')

plt.show()
```

```python
trafico = df['Traffic_Density'].value_counts()
trafico.plot(kind='bar')
plt.title('Densidad del tráfico')

plt.show()
```

```python
felicidad = df['Happiness_Score'].value_counts()
felicidad.plot(kind='bar')
plt.title('Distribución del Happiness Score')
plt.xlabel('Nivel de felicidad')


plt.show()
```

```python
costo= df['Cost_of_Living_Index'].value_counts()
plt.figure(figsize=(12, 12))
plt.pie(costo, labels=costo.index, autopct='%1.1f%%', startangle=140, colors=['#9999ff', '#ff9999'])
plt.title('Costo de vida')

plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(df['Year'],df['Month'], alpha=0.5, c=(df['Decibel_Level']), cmap='viridis')
plt.xlabel('Ciudad')
plt.ylabel('Decibelios')
plt.title(' Dispersión: Ruido por mes ')
color_bar = plt.colorbar()



plt.show()
```

```python

intervalos = np.arange(0, 500, 50) 
frecuencias, bordes = np.histogram(df['Green_Space_Area'], bins=intervalos)
plt.bar(range(len(frecuencias)), frecuencias, color='skyblue')
plt.xlabel('Intervalos de Área de Espacio Verde')
plt.ylabel('Frecuencia')
plt.xticks(range(len(intervalos)), intervalos)

plt.show()
```

```python
sns.countplot(x='City', hue=intervalos.size, data=df, palette='Set1')
plt.title('Distribución de Área de Espacio Verde por Ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Conteo')
plt.legend(title='Green_Space_Area')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x='Healthcare_Index', y='Air_Quality_Index', data=df, errorbar=None)
plt.title('Relación entre sanidad y aire')
plt.xlabel('Health')
plt.ylabel('Air')


plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x='Cost_of_Living_Index', y='Happiness_Score', data=df, errorbar=None)
plt.title('Relación entre costo y felicidad')
plt.xlabel('Costo')
plt.ylabel('Felicidad')


plt.show()
```
